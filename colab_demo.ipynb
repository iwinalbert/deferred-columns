### Step 3: Load the Extension & Test Lazy Materialization
We will load our compiled `.duckdb_extension` file and query our `deferred_scan` Table Function.

```
con = duckdb.connect()
# Load the compiled extension
con.execute("LOAD 'build/release/extension/deferred_columns/deferred_columns.duckdb_extension';")

# Test the Deferred Scan
print("Deferred Scan Output:")
display(con.execute("SELECT * FROM deferred_scan('mock_model');").df())
```

### Step 4: Test Approximate Query Processing (AQP)
Now we test the `sum_ci` aggregate function. It natively calculates the sum and a 95% Confidence Interval based on a provided RMSE (e.g. 0.5).

```
print("\nAQP Sum with Confidence Intervals:")
display(con.execute("SELECT sum_ci(imputed_value, 0.5) AS bounded_sum FROM deferred_scan('mock_model');").df())
```
